Data Cleaning

```python
car_rental.isna().value_counts()
```

```python
a=car_rental.duplicated()
print(a.sum())
```

```python
null_rows = car_rental[car_rental.isnull().any(axis=1)]
null_rows
```

```python
categories=car_rental.select_dtypes(include=['str']).columns
categories
```

```python
car_rental[['Brand', 'Model', 'Fuel_Type', 'Transmission', 'Body_Type']].isnull().sum()
```

```python
for i in categories:
    car_rental[i]=car_rental[i].fillna("unknown",inplace=True)
```

```python
car_rental.isna().sum()
car_rental["Brand"].unique()
```

```python
numeric=car_rental.select_dtypes(include=['float'])
numeric
```

```python
for i in numeric:
    car_rental[i]=car_rental[i].fillna(car_rental[i].median())
```

```python
car_rental.isna().sum()
```

```python
car_rental.to_csv(
    "../backend/dataset/car_rental_cleaned.csv",
    index=False
)
```
