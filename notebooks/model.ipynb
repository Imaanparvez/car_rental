Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def build_user_profile(user_preferences, tfidf):
    """    Converts user-selected categorical preferences into a TF-IDF vector
    """
    user_text = (
        user_preferences["Brand"] + " " +
        user_preferences["Fuel_Type"] + " " +
        user_preferences["Transmission"] + " " +
        user_preferences["Body_Type"]
    )
    
    user_vector = tfidf.transform([user_text])
    return user_vector
```

```python
def apply_numeric_filters(df1, preferences):
    """
    Apply numeric filters: Mileage and Engine_CC only
    """

    filtered_df = df1.copy()

    # Minimum mileage
    if "min_mileage" in preferences and preferences["min_mileage"] is not None:
        filtered_df = filtered_df[
            filtered_df["Mileage"] >= preferences["min_mileage"]
        ]

    # Maximum engine capacity
    if "max_engine_cc" in preferences and preferences["max_engine_cc"] is not None:
        filtered_df = filtered_df[
            filtered_df["Engine_CC"] <= preferences["max_engine_cc"]
        ]

    return filtered_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_cbf(
    user_preferences,
    df1,
    car_tfidf_matrix,
    tfidf,
    top_n=5
):
    # 1. Build user preference text
    user_text = (
        user_preferences["Brand"] + " " +
        user_preferences["Fuel_Type"] + " " +
        user_preferences["Transmission"] + " " +
        user_preferences["Body_Type"]
    )

    user_vector = tfidf.transform([user_text])

    # 2. Compute cosine similarity
    similarity_scores = cosine_similarity(
        user_vector,
        car_tfidf_matrix
    )[0]

    df1 = df1.copy()
    df["similarity_score"] = similarity_scores

    # 3. Apply numeric filters (Mileage + CC)
    df_filtered = apply_numeric_filters(df1, user_preferences)

    # 4. Rank and return Top-N
    recommendations = df_filtered.sort_values(
        by="similarity_score",
        ascending=False
    ).head(top_n)

    return recommendations[[
        "Car_ID",
        "Brand",
        "Model",
        "Fuel_Type",
        "Transmission",
        "Body_Type",
        "Mileage",
        "Engine_CC",
        "similarity_score"
    ]]
```
