```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../backend/dataset/car_rental_cleaned.csv")
df
```

```python
df.isnull().sum()
```

Category Visualisation

```python
df["Brand"].value_counts().plot(kind="bar")
plt.title(" Car Brands")
plt.xlabel("Brand")
plt.ylabel("Number of Cars")
plt.show()
```

```python
df["Body_Type"].value_counts().plot(kind="bar")
plt.title("Car Body Type Distribution")
plt.xlabel("Body Type")
plt.ylabel("Count")
plt.show()
```

```python
df["Fuel_Type"].value_counts().plot(kind="bar")
plt.title("Fuel Type Distribution")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.show()

```

Numerical Features

```python
plt.hist(df["Mileage"], bins=30)
plt.title("Mileage Distribution")
plt.xlabel("Mileage")
plt.ylabel("Frequency")
plt.show()

plt.hist(df["Engine_CC"], bins=30)
plt.title("Engine CC Distribution")
plt.xlabel("Engine CC")
plt.ylabel("Frequency")
plt.show()

df["Year"].value_counts().sort_index().plot(kind="line")
plt.title("Car Manufacturing Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

Outliers

```python
plt.boxplot(df["Mileage"])
plt.title("Mileage Outliers")
plt.ylabel("Mileage")
plt.show()

plt.boxplot(df["Engine_CC"])
plt.title("Engine CC Outliers")
plt.ylabel("Engine CC")
plt.show()
```

```python
plt.scatter(df["Mileage"], df["Engine_CC"])
plt.title("Mileage vs Engine CC")
plt.xlabel("Mileage")
plt.ylabel("Engine CC")
plt.show()
```
